## Toposort与DFS

用的是书P45 quiz8.3的图做的

```python
def DFS(data,startpoint,exploredlist,curlabel,orders):
    exploredlist.append(startpoint)
#     print(exploredlist)
    linkedlist=data[:,1][np.where(data[:,0]==startpoint)]
#     print(linkedlist)
    for point in linkedlist:
        if not (point in exploredlist):
            orders,curlabel,exploredlist=DFS(data,point,exploredlist,curlabel,orders)
    orders[int(startpoint-1)]=curlabel
    curlabel=curlabel-1
    return orders,curlabel,exploredlist
def Toposort(data):
    orders=np.zeros([int(data.max()),1])
    curlabel=int(data.max())
    exploredlist=[]
    for startpoint in range(1,int(data.max())+1):
        if not (startpoint in exploredlist):
            orders,curlabel,exploredlist=DFS(data,startpoint,exploredlist,curlabel,orders)
    return orders
```

## Kosaraju

```python
def Kosaraju(data):
    datarev=np.zeros(data.shape)
    datarev[:,0]=data[:,1]
    datarev[:,1]=data[:,0]
    initialorders=Toposort(datarev)
    exploredlist=[]
    curlabel=int(data.max())
    orders=np.zeros([int(data.max()),1])
    numSCC=[]
    for vertexidx in range(1,int(data.max())+1):
        vertex=(np.where(initialorders==vertexidx))[0]+1#按照顺序取出来各个点
        if not vertex in exploredlist:
            orders,curlabel,exploredlist=DFS(data,vertex,exploredlist,curlabel,orders)
            if np.count_nonzero(orders)>1:
                numSCC.append(np.count_nonzero(orders))
            orders=np.zeros([int(data.max()),1])
    numSCC.sort(reverse=True)
    if len(numSCC)<5:
        while len(numSCC) !=5:
            numSCC.append(0)
    elif len(numSCC)>5:
        while len(numSCC) !=5:
            del numSCC[-1]
    return numSCC
```

# assignment2Dijkstra

```python
def matrixgraph(f):
    #把原始数据用矩阵表示
    df=pd.DataFrame(data=np.zeros([200,200]),index=range(1,201),columns=(range(1,201)))
    for line in f:
        line=line.split('\t')
        rawnumber=int(line[0])
        for pointidx in range(1,len(line)):
            point=line[pointidx].split(',')
#             print(point)
            columnnumber=int(point[0])
            distance=int(point[1])
#             print(df.iloc[rawnumber-1,columnnumber-1])
            if (df.iloc[rawnumber-1,columnnumber-1] == 0 or df.iloc[rawnumber-1,columnnumber-1]>distance):
                df.iloc[rawnumber-1,columnnumber-1]=distance
    df=df.replace(0,np.nan)
    return df
```

```python
def dijkstra(f):
    path={}
    path[1]=[1]
    df=matrixgraph(f)
    X=[1]
    distancev=np.zeros(200)
    searchcolumn=[i for i in df.columns if i not in X]
    searchdf=df.loc[X][searchcolumn]
    newpoint=int(searchdf.idxmin(axis=1))
    X.append(newpoint)
    path[newpoint]=[1,newpoint]
    distancev[newpoint-1]=searchdf.min(axis=1)
    searchcolumn=[i for i in df.columns if i not in X] #还没有explored的点
    searchdf=df.loc[X][searchcolumn]
    for raw in X:
        searchdf.loc[raw]=searchdf.loc[raw]+int(distancev[raw-1]) 
    while not pd.isnull(searchdf.min().min()):
        #确定出表里面最小的点以及列号就是要加进去的点了
        newpoint=int((searchdf.min(axis=0)).idxmin())
        X.append(newpoint)
        linkpoint=int((searchdf.min(axis=1)).idxmin())
        path[newpoint]=path[linkpoint].copy()
        path[newpoint].append(newpoint)
        distancev[newpoint-1]=searchdf.min().min()
        searchcolumn=[i for i in df.columns if i not in X] #还没有explored的点
        searchdf=df.loc[X][searchcolumn]

        for raw in X:
            searchdf.loc[raw]=searchdf.loc[raw]+int(distancev[raw-1]) 

    answer=[distancev[i] for i in [6,36,58,81,98,114,132,164,187,196]]
    return answer,path
```

# assignment3Median

```python
def Median(data):
    import heapq as heapq
    sum=data[0]
    h1=[-min(data[0],data[1])]
    sum=sum-h1[0]
    heapq.heapify(h1)
    h2=[max(data[0],data[1])]
    heapq.heapify(h2)
    for i in range(2,len(data)):
    #     print(data[i])
        if data[i]>-h1[0]:
            heappush(h2,data[i])
        else:
            heappush(h1,-data[i])
        while len(h1)-len(h2)>1:
            heappush(h2,-h1[0])
            heappop(h1)
        while len(h2)>len(h1):
            heappush(h1,-h2[0])
            heappop(h2)  
        m=-h1[0]
        sum=sum+m
        
    return sum%10000
```

# assignment4TwoSum

```python

def sumhash(A,t):
    H={}
    for i in (range(len(A))):
        H[A[i]]=i
    for i in (range(len(A))):
        y=t-A[i]
        if y in H:
            return True
    return False
def TwoSumfinal(A):
    sum=0  
    for t in range(-10000,10001):
        answer=sumhash(A,t)
        if answer:
            sum=sum+1
    return sum
```
