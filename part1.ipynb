```python
def Karatsuba(n1,n2):
    if n1>9 and n2>9:
        len1=int(math.log10(n1)+1)
        len2=int(math.log10(n2)+1)
        cuti1=int(len1/2)
        cuti2=int(len2/2)
        a=int(str(n1)[:cuti1])
        b=int(str(n1)[cuti1:])
        c=int(str(n2)[:cuti2])
        d=int(str(n2)[cuti2:])
        out=10**(len1+len2-cuti1-cuti2)*Karatsuba(a,c)+10**(cuti1)*(Karatsuba((a+b),(c+d))-Karatsuba(a,c)-Karatsuba(b,d))+Karatsuba(b,d)  
    else:
#         print(n1)
#         print(n2)
        out=n1*n2
    return out
```

```python
def partition(A,l,r):
    #l是起始值0，r是最终值，如len(A)-1
    comparecount=0
    p=A[l]
    i=l+1
    for j in range (i,r+1):#注意如果遍历到最后一个数需要到r+1
#         print(j)

        comparecount=comparecount+1
        if A[j]<p:
            temp=A[i]
            A[i]=A[j]
            A[j]=temp
            i=i+1
    temp=A[l]
    A[l]=A[i-1]
    A[i-1]=temp
    return i-1,A,comparecount
# A=[1,2,5,9,7,8,100,135,11,9999]
```

```python
def choosepivotfirst(A,l,r):
    return l
def leftquicksort(A,l,r):
    count=0
    if l>=r:
        return A,count
    else:
        i=choosepivotfirst(A,l,r)
#         print(i)
        temp=A[l]
        A[l]=A[i]
        A[i]=temp
        j,A,count1=partition(A,l,r)
#         print(j)
        A,count2=leftquicksort(A,l,j-1)
        A,count3=leftquicksort(A,j+1,r)
        return A,count1+count2+count3
```

```python
def choosepivotmiddle(A,l,r):
    if (r-l+1) % 2 == 0:
        k=int((r-l+1)/2)-1+l
    else:
        k=int(len(A[l:r])/2)+l
    B=[A[l],A[k],A[r]]
    B,countnouse=leftquicksort(B,0,2)
#     print(B)
    for i in [l,k,r]:
        if A[i]==B[1]:
            out=i
    return out

# test median

def quicksort(A,l,r):
    count=0
    if l>=r:
        return A,count
    else:
        i=choosepivotmiddle(A,l,r)
#         print(i)
#         print(i)
        temp=A[l]
        A[l]=A[i]
        A[i]=temp
        j,A,count1=partition(A,l,r)
#         print(j)

        A,count2=quicksort(A,l,j-1)
        A,count3=quicksort(A,j+1,r)
        return A,count1+count2+count3
# A=objectarray
# print(quicksort(A,0,len(A)-1) )
A=[7, 1, 3, 6, 2, 5, 4, 9, 8]
A[choosepivotmiddle(A,0,len(A)-1)]
A=[4, 1, 3, 6, 2, 5]
A[choosepivotmiddle(A,0,len(A)-1)]
```

```python
def contraction(A,sizeA):
# 输入原始矩阵，只进行一次随机运算，给出来最后还剩下的edge总数
#     print(A)#打印原始数据
#     print('sizeA'+str(sizeA))
    for iteration in range (A.shape[0]-2):#最后只剩两个数
#     for iteration in range (1):
#         print('newloop')
    
        i,j=choose(A)# 选出来任意一个点
#         np.savetxt('Asave.txt',A)
#         print(A[i,j])
        cleanedone=int(A[i,j]-1)# 所选点列号对应的vector删掉，行保留
        retainedone=int(i)
#         print('cleanedone'+str(cleanedone))
#         print('retainedone'+str(retainedone))
        k=1# 把删除的vector连接的所有点合并过去
        while k != A.shape[1]:# 把保留点中所有和clean点的连线都删除
            if A[retainedone,k]==cleanedone+1:
                A[i,k:(A.shape[1]-1)]=A[i,k+1:]
                k=k-1
                sizeA[i]=sizeA[i]-1
            k=k+1
#         print('cleandedA')
#         print(A)
#         print('sizeA'+str(sizeA))
        for it2 in range(1,int(sizeA[cleanedone])):# 每一行第一个数是行号，所以检索的时候也从第二列开始
            if A[cleanedone,it2] != retainedone+1: # cleandedone里面，除了保留点本身的连线不加，其他连线全都合并到保留点里面  ，即便clean点和保留点之前有共同的连线点也要加上， 因为那些连接没有被删除，而是出现在合并点中

                sizeA[retainedone]=sizeA[retainedone]+1

                A[retainedone,int(sizeA[retainedone]-1)]=A[cleanedone,it2]
        A[cleanedone,:]=np.zeros(A.shape[1]) #为了保持行号还是正确的，所以不删除行只是全部清除为0
        sizeA[cleanedone]=0
        A[np.where(A==cleanedone+1)]=retainedone+1#把别的点和被清除点的连线全都改掉
#         print('switchedA'+str(A))
    return sizeA[np.where(sizeA!=0)][0]-1# size本身算了行号
    
def choose(A):
    i=random.randint(0, A.shape[0]-1)
    j=random.randint(1,A.shape[1]-1) #选的时候要跳过行号，否则就是自己和自己配对了
    while A[i,j]==0:
        i=random.randint(0, A.shape[0]-1)
        j=random.randint(1,A.shape[1]-1)
    return i,j
```

```python
def contraction(A,sizeA):
# 输入原始矩阵，只进行一次随机运算，给出来最后还剩下的edge总数
    print(A)#打印原始数据
    print('sizeA'+str(sizeA))
    for iteration in range (A.shape[0]-2):#最后只剩两个数
#     for iteration in range (1):
        print('newloop')
    
        i,j=choose(A)# 选出来任意一个点
#         np.savetxt('Asave.txt',A)
#         print(A[i,j])
        cleanedone=int(A[i,j]-1)# 所选点列号对应的vector删掉，行保留
        retainedone=int(i)
        print('cleanedone'+str(cleanedone))
        print('retainedone'+str(retainedone))
        k=1# 把删除的vector连接的所有点合并过去
        while k != A.shape[1]:# 把保留点中所有和clean点的连线都删除
            if A[retainedone,k]==cleanedone+1:
                A[i,k:(A.shape[1]-1)]=A[i,k+1:]
                k=k-1
                sizeA[i]=sizeA[i]-1
            k=k+1
        print('cleandedA')
        print(A)
        print('sizeA'+str(sizeA))
        for it2 in range(1,int(sizeA[cleanedone])):# 每一行第一个数是行号，所以检索的时候也从第二列开始
            if A[cleanedone,it2] != retainedone+1: # cleandedone里面，除了保留点本身的连线不加，其他连线全都合并到保留点里面  ，即便clean点和保留点之前有共同的连线点也要加上， 因为那些连接没有被删除，而是出现在合并点中
#                print(A[cleanedone,it2])
                sizeA[retainedone]=sizeA[retainedone]+1
#                 print(sizeA)
#                 print(A)
                A[retainedone,int(sizeA[retainedone]-1)]=A[cleanedone,it2]
        A[cleanedone,:]=np.zeros(A.shape[1]) #为了保持行号还是正确的，所以不删除行只是全部清除为0
        sizeA[cleanedone]=0
        A[np.where(A==cleanedone+1)]=retainedone+1#把别的点和被清除点的连线全都改掉
        print('switchedA'+str(A))
    return sizeA[np.where(sizeA!=0)][0]-1# size本身算了行号
    
def choose(A):
    i=random.randint(0, A.shape[0]-1)
    j=random.randint(1,A.shape[1]-1) #选的时候要跳过行号，否则就是自己和自己配对了
    while A[i,j]==0:
        i=random.randint(0, A.shape[0]-1)
        j=random.randint(1,A.shape[1]-1)
    return i,j
```
